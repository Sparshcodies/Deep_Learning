```python
# sigmoid activation function
def sigmoid_numpy(x):
  return 1/(1+np.exp(-x))
```

```python
# log loss or binary loss function

def log_loss(y, yhat):
  e = 1e-10
  yhat_new = [max(i,e) for i in yhat]
  yhat_new = [min(i,1-e) for i in yhat_new]
  yhat_new = np.array(yhat_new)
  return -np.mean(y*np.log(yhat_new)+(1-y)*np.log(1-yhat_new))
```

```python
def gradient_descent(age, affordability, y_true, epochs, loss_thresold):
    w1 = w2 = 1
    bias = 0
    rate = 0.5
    n = len(age)
    for i in range(epochs):
        weighted_sum = w1 * age + w2 * affordability + bias
        y_predicted = sigmoid_numpy(weighted_sum)
        loss = log_loss(y_true, y_predicted)

        w1d = (1/n)*np.dot((age),(y_predicted-y_true))
        w2d = (1/n)*np.dot((affordability),(y_predicted-y_true))
        #print("weighted_sum",weighted_sum, "y_predicted",y_predicted,"loss",loss,"w1d",w1d,"w2d",w2d)
        bias_d = np.mean(y_predicted-y_true)

        w1 = w1 - rate * w1d
        w2 = w2 - rate * w2d
        bias = bias - rate * bias_d

        print (f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')

        if loss<=loss_thresold:
            break

    return w1, w2, bias

X_train_age = np.array(X_train['age'])
X_train_affordability =np.array(X_train['affordibility'])
w_1, w_2, b = gradient_descent(X_train_age,X_train_affordability,y_train,120,0.60)
```

```python
def sigmoid(x):
  return 1/(1+math.exp(-x))
```

```python
# log loss or binary loss function

def log_loss_3(y, yhat):
  e = 1e-10
  yhat_new =max(yhat,e)
  yhat_new = min(yhat_new,1-e)

  return -(y*np.log(yhat_new)+(1-y)*np.log(1-yhat_new))
```

```python

def stochastic_gradient_descent(age, affordability, y_true, epochs, loss_threshold):
    w1 = w2 = 1
    bias = 0
    rate = 0.5
    n = len(age)

    for epoch in range(epochs):
        for i in range(n):
            index = np.random.randint(0, n-1)
            weighted_sum = w1 * age[index] + w2 * affordability[index] + bias
            y_predicted = sigmoid(weighted_sum)

            loss = log_loss_3(y_true[index], y_predicted)

            w1d = age[index] * (y_predicted - y_true[index])
            w2d = affordability[index] * (y_predicted - y_true[index])
            bias_d = (y_predicted - y_true[index])

            w1 = w1 - rate * w1d
            w2 = w2 - rate * w2d
            bias = bias - rate * bias_d

        print(f'Epoch:{epoch}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')

    return w1, w2, bias



X_train_age = np.array(X_train['age'])
X_train_affordability =np.array(X_train['affordibility'])
y_train_new = np.array(y_train)

stochastic_gradient_descent(X_train_age,X_train_affordability, y_train_new, 100, 0.4631)
```

```python

def mini_batch_gradient_descent(age, affordability, y_true, epochs, loss_threshold, batch_size=32):
    w1 = w2 = 1
    bias = 0
    rate = 0.5
    n = len(age)

    for i in range(epochs):
        for j in range(0, n, batch_size):
            age_batch = age[j:j+batch_size]
            affordability_batch = affordability[j:j+batch_size]
            y_true_batch = y_true[j:j+batch_size]

            weighted_sum = w1 * age_batch + w2 * affordability_batch + bias
            y_predicted = sigmoid_numpy(weighted_sum)
            loss = log_loss(y_true_batch, y_predicted)

            w1d = (1/batch_size) * np.dot(age_batch, np.transpose(y_predicted - y_true_batch))
            w2d = (1/batch_size) * np.dot(affordability_batch, np.transpose(y_predicted - y_true_batch))
            bias_d = np.mean(y_predicted - y_true_batch)

            w1 = w1 - rate * w1d
            w2 = w2 - rate * w2d
            bias = bias - rate * bias_d

        print(f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')

        if loss <= loss_threshold:
            break

    return w1, w2, bias

mini_batch_gradient_descent(X_train_age,X_train_affordability, y_train, 100, 0.4631, batch_size=10)
```
