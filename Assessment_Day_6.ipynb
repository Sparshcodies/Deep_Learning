```python
data.replace({"Female":1,"Male":0},inplace=True)
```

```python
data_pro=pd.get_dummies(data=data,columns=["BusinessTravel","Department","EducationField","JobRole","MaritalStatus"])
```

```python
col_to_scale=["DailyRate","Age","DistanceFromHome","Education","HourlyRate","HourlyRate","EnvironmentSatisfaction","JobInvolvement","JobLevel","JobSatisfaction","MonthlyIncome","MonthlyRate","NumCompaniesWorked","PercentSalaryHike","RelationshipSatisfaction","Shift","TotalWorkingYears","TrainingTimesLastYear","WorkLifeBalance","YearsAtCompany","YearsInCurrentRole","YearsSinceLastPromotion","YearsWithCurrManager",]
scaler=MinMaxScaler()
data_pro[col_to_scale]=scaler.fit_transform(data_pro[col_to_scale])
```

```python
col_unique_value(data_pro)
```

```python
data_pro
```

```python
data_pro
```

```python
X=data_pro.drop("Attrition",axis=1)
Y=data_pro["Attrition"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=2, train_size = .75)
```

```python
print(type(X_train))
print(X_train.shape)
```

```python
model=keras.Sequential([
                        keras.layers.Dense(100,input_shape=(44,),activation="sigmoid"),
                        keras.layers.Dense(50,activation="relu"),
                        keras.layers.Dense(2,activation="sigmoid"),
                        ])
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
model.fit(X_train,y_train,epochs=200)
```

```python
model.evaluate(X_test,y_test)
```

```python
ans=model.predict(X_test)
```

```python
ans
```

```python
final_ans=[np.argmax(i) for i in ans]
final_ans=np.array(final_ans)
```
